## Identify Outliers and perform outliers technique

```python
#sns.boxplot(x=df["danceability"]) 

#sns.boxplot(x=df["duration_minutes"])
```

```python
# danceability
# Winsorization

df["danceability"].describe()

Q1 = df["danceability"].quantile(0.25)
Q3 = df["danceability"].quantile(0.75)
IQR = Q3-Q1
UL = Q3+1.5*IQR
LL = Q1-1.5*IQR
LL,UL
```

```python
outlier_data = df[(df["danceability"]<LL) | (df["danceability"]>UL)]
outlier_data
outlier_data.describe()
```

```python
# fill the outliers with upper and lower limit

row_index = df["danceability"].index
for i in row_index:
    if df.loc[i,"danceability"] < LL:
        df.loc[i,"danceability"] = LL
    elif df.loc[i,"danceability"] > UL:
        df.loc[i,"danceability"] =UL
        
#sns.boxplot(df["danceability"]) # checking if there are outliers
```

```python
#duration_minutes
df["duration_minutes"].describe()

Q1 = df["duration_minutes"].quantile(0.25)
Q3 = df["duration_minutes"].quantile(0.75)
IQR = Q3-Q1
UL = Q3+1.5*IQR
LL = Q1-1.5*IQR
LL,UL
```

```python
outlier_data_min = df[(df["duration_minutes"]<LL) | (df["duration_minutes"]>UL)]
outlier_data_min
outlier_data_min.describe()
```

```python
# fill the outliers with upper and lower limit

row_index = df["duration_minutes"].index
for i in row_index:
    if df.loc[i,"duration_minutes"] < LL:
        df.loc[i,"duration_minutes"] = LL
    elif df.loc[i,"duration_minutes"] > UL:
        df.loc[i,"duration_minutes"] =UL
        
#sns.boxplot(df["duration_minutes"]) # checking if there are outliers
```
